```python
import matplotlib.pyplot as plt


%matplotlib inline 

import numpy as np
```

```python
x=np.linspace(0,15)
y= x **2
plt.plot(x,y)
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r--')

plt.subplot(1,3,2)
plt.plot(y,x,'g')

plt.subplot(1,3,3)
plt.plot(y,y,'p--')
```

```python
fig= plt.figure()

axes= fig.add_axes([0.1,0.1,1,1])

axes.plot(x,y,'g',label='YOLO')

axes.legend()
```
